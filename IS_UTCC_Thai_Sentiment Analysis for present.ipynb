# การหาคุณลักษณะพิเศษ (Feature Extraction)

### Word Tokenize

```python
# ทำการเรียก Module ตัดคำภาษาไทย
from pythainlp import word_tokenize
```

```python
# สร้าง Function ในการตัดคำ และ ทำความสะอาดข้อมูล
def get_th_token(text):
    text = text.lower() 
    text = text.replace('\n',' ')
    token = word_tokenize(text, engine="newmm")
    return token
```

```python
# ทดสอบ Function : get_th_token
samp_text='เเพ็คห่อมาเเน่นหนาดีมาก'
get_th_token(samp_text)
```

### Bag of Words

# สร้างแบบจำลอง(Model) และการทดสอบประสิทธิภาพ

### LinearSVM

```python
from sklearn.svm import LinearSVC

from sklearn.model_selection import cross_validate, cross_val_score
from sklearn.metrics import (classification_report, make_scorer, 
                             confusion_matrix, accuracy_score, 
                             precision_score, recall_score, f1_score)
```

```python
model_SVC = LinearSVC(random_state=1987)
```

```python
model_SVC.fit(train_feature_normalize, train_label)
```

```python
c5_SVC = cross_val_score(model_SVC, train_feature_normalize, 
                         train_label, cv=5, scoring='accuracy')
mc5_SVC = cross_val_score(model_SVC, train_feature_normalize, 
                          train_label, cv=5, scoring='accuracy').mean()
print(c5_SVC)
print(mc5_SVC)
```

```python
scoring = {'accuracy' : make_scorer(accuracy_score), 
           'precision' : make_scorer(precision_score),
           'recall' : make_scorer(recall_score), 
           'f1_score' : make_scorer(f1_score)}

c5_SVC_CV_score = cross_validate(model_SVC, train_feature_normalize, 
                                 train_label, cv=5, scoring=(scoring))
c5_SVC_CV_score
```

```python
c5_SVC_CV_score_table = pd.DataFrame(c5_SVC_CV_score)
c5_SVC_CV_score_table
```

```python
c5_SVC_CV_score_board = c5_SVC_CV_score_table[['test_accuracy', 
                                               'test_precision', 
                                               'test_recall', 
                                               'test_f1_score']].mean()
c5_SVC_CV_score_board
```

```python
from yellowbrick.model_selection import CVScores

visualizer_c5_SVC = CVScores(model_SVC, cv=5, scoring='accuracy')

visualizer_c5_SVC.fit(train_feature_normalize, train_label)
visualizer_c5_SVC.show() 
```

```python
c10_SVC = cross_val_score(model_SVC, train_feature_normalize, 
                          train_label, cv=10, scoring='accuracy')
mc10_SVC = cross_val_score(model_SVC, train_feature_normalize, 
                           train_label, cv=10, scoring='accuracy').mean()
print(c10_SVC)
print(mc10_SVC)
```

```python
scoring = {'accuracy' : make_scorer(accuracy_score), 
           'precision' : make_scorer(precision_score),
           'recall' : make_scorer(recall_score), 
           'f1_score' : make_scorer(f1_score)}

c10_SVC_CV_score = cross_validate(model_SVC, train_feature_normalize, 
                                  train_label, cv=10, scoring=(scoring))
c10_SVC_CV_score
```

```python
c10_SVC_CV_score_table = pd.DataFrame(c10_SVC_CV_score)
c10_SVC_CV_score_table
```

```python
predict_SVC = model_SVC.predict(test_feature_normalize)
```

```python
acc_SVC = classification_report(test_label, predict_SVC, digits=4)
print(acc_SVC)
```

```python
con_mat_SVC = confusion_matrix(test_label, predict_SVC)
print(con_mat_SVC)
```

### 5.2 อภิปรายผลการทดลอง

5.2.1 Engine ในการตัดคำจากโมดูล PythaiNLP 2.0.7
	การตัดคำ (Word Tokenize) ส่งผลต่อค่าประสิทธิภาพของแบบจำลอง (Model) สำหรับโมดูลในการตัดคำจาก PythaiNLP 2.0.7 จะให้ค่าเริ่มต้น (default) ของ engine = “newmm” แต่เมื่อได้ทำการทดลองด้วย engine อื่นๆ เช่น “ulmfit”  และ “longest” จะให้ค่าประสิทธิภาพที่เปลี่ยนแปลงไป แต่ engine = “newmm” จะให้ผลลัพธ์ที่ดีที่สุด ถ้าหากพิจารณา ค่าความถูกต้อง (Accuracy) จากอัลกอริทึมซัพพอร์ทเวคเตอร์แมชชีน จาการเลือก engine ในการตัดคำเป็น  “newmm” “ulmfit”  และ “longest” จะได้ค่าความถูกต้อง เท่ากับ 91.20% 91.20% และ 89.20% ตามลำดับ

5.2.2 Normalize Feature Matrix
	การ Normalize Feature Matrix ส่งผลต่อค่าส่งผลต่อค่าประสิทธิภาพของแบบจำลอง (Model) สำหรับการทดลองนี้  ใช้วิธีการ L2 Normalize จะส่งผลให้ค่าประสิทธิภาพเพิ่มขึ้นหากเปรียบเทียบกับการไม่ทำ Normalize ถ้าหากพิจารณา ค่าความถูกต้อง (Accuracy) จากอัลกอริทึมซัพพอร์ทเวคเตอร์แมชชีน ด้วยการทำ L2 Normalize และ ไม่ทำ Normalize จะได้ค่าความถูกต้อง เท่ากับ 91.20% และ 90.40% ตามลำดับ
